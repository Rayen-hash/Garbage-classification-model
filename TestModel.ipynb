```python
import cv2
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt

CATEGORIES = ["battery","biological","cardboard","clothes","glass","metal","paper","plastic","shoes"]
test_path = "C:/GLSI3/tp_machine/Test_images"
IMG_SIZE = 80

def prepare(filepath):
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

model = tf.keras.models.load_model("model_1.h5")

for img_name in os.listdir(test_path):
    full_path = os.path.join(test_path, img_name)
    img = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)
    prediction = model.predict(prepare(full_path))
    class_idx = np.argmax(prediction)
    predicted_label = CATEGORIES[class_idx]

    # Affichage avec matplotlib
    plt.imshow(img , cmap="gray")
    plt.title(f"Prediction: {predicted_label}")
    plt.axis('off')
    plt.show()



```
